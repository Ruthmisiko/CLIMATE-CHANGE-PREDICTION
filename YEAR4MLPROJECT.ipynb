
# <span style="background-color:lightyellow; padding:10px;">JOINING DATASET FOR ALL 47 COUNTIES</span>


```python
import os
import pandas as pd

data_directory = '/home/ruth/CLIMATE CHANGE'
counties = [
    'BARINGO', 'BOMET', 'BUNGOMA', 'BUSIA', 'ELGEYO MARAKWET', 'EMBU', 'GARISSA',
    'HOMABAY', 'ISIOLO', 'KAJIADO', 'KAKAMEGA', 'KERICHO', 'KIAMBU', 'KILIFI',
    'KIRINYAGA', 'KISII', 'KISUMU', 'KITUI', 'KWALE', 'LAIKIPIA', 'LAMU', 'MACHAKOS',
    'MAKUENI', 'MANDERA', 'MARSABIT', 'MERU', 'MIGORI', 'MOMBASA', 'MURANGA', 'NAIROBI',
    'NAKURU', 'NANDI', 'NAROK', 'NYAMIRA', 'NYANDARUA', 'NYERI', 'SAMBURU', 'SIAYA',
    'TAITA TAVETA', 'TANA RIVER', 'THARAKA', 'TRANSNZOIA', 'TURKANA', 'UASIN GISHU',
    'VIHIGA', 'WAJIR', 'WEST POKOT'
]  

dfs = []

for county in counties:
    file_path = os.path.join(data_directory, f'{county}.csv')
    
    # data for each county, skipping the first 11 rows as headers
    df = pd.read_csv(file_path, skiprows=11)
    
    # Additional column for county to identify data
    df['County'] = county
    
    dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)

final_df.to_csv('/home/ruth/CLIMATE CHANGE/climatedata.csv', index=False)
```
